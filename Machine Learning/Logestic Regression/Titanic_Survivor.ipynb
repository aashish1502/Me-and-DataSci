```python
print("The average age for 1st class is",train[train["Pclass"]==1]["Age"].mean())
print("The average age for 2nd class is",train[train["Pclass"]==2]["Age"].mean())
print("The average age for 3rd class is",train[train["Pclass"]==3]["Age"].mean())
```

```python
def age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 25
    else:
        return Age
        
```

```python
train['Age']= train[["Age","Pclass"]].apply(age,axis =1)
```

```python
train
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
train.drop("Cabin",axis=1,inplace=True)
```

```python
train
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
 #Dummy variable
sex = pd.get_dummies(train["Sex"],drop_first=True)
```

```python
sex.head()
```

```python
embark =pd.get_dummies(train["Embarked"],drop_first=True)
```

```python
embark.head()
```

```python
train=pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

```python
train.drop(["Sex","Embarked","Name","Ticket"],axis=1,inplace=True)
```

```python
train.head()
```

```python
train.drop(["PassengerId"],axis=1,inplace=True)
```

```python
train.head()
```

```python
y=train["Survived"]
```

```python
X=train.drop(["Survived"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
