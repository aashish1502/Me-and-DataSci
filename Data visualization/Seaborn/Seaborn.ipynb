# How to use seaborn

```
import seaborn as sns
```

```
%matplotlib inline
```

##Loading a precreated data set by using the sns.load_dataset() function

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

## To check the distrubution of a univarient and plot it on a histogram we use the sns.distplot()

```
sns.distplot(tips['total_bill'] , kde = False, bins = 50)
# KDE is kernal density estimation and it is a summation of the waves 
# bins are the number at which we jump the data point
```

## Joint plot 

```
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
# the scatter plot is the amulgamation of the frequency according to the values of both variables
# the kind arg styles the scatterplot
```

```
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')

# the scatter plot is the amulgamation of the frequency according to the values of both variables
# the kind arg styles the scatterplot
```

```
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')

# the scatter plot is the amulgamation of the frequency according to the values of both variables
# the kind arg styles the scatterplot
```

## sns.pairplot(dataframe) plots pairwise relationship with every possible combination

```
sns.pairplot(tips)
# pairplot is a very nice way to quickly visualize your data
```

```
sns.pairplot(tips,hue='sex')
#The hue refines the plots according to a classification refrencing
```

```
sns.pairplot(tips,palette ='coolwarm',hue='sex')
```
