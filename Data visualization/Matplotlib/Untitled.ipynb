## There are two ways to plot a series of data in matplot lib one is by using functions and the other is by using objects

```python
#Function oriented (less recommnended)
plt.plot(x,y)
```

### To label the axis we use

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y label")
plt.title("Title")
```

### To plot multiple plots we use

```python
#plt.subplot(NumberOfRows,NumberOfColumns,TheColumnNumber)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object oriented method

```python
# The code bellow creates a figure object
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

### The function figure() creates a blank canvas on which we can do futher justifications for out plot

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([.2,.5,.4,.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,.5,.5])
axes2 = fig.add_axes([0,.6,.5,.5])
axes3 = fig.add_axes([.6,.6,.5,.5])
axes4 = fig.add_axes([.6,0,.5,.5])
```

 ### How to create sub plots

```python
fig,axes = plt.subplots(nrows = 3 , ncols = 3)
plt.tight_layout()
```

```python
figs,axes  = plt.subplots(nrows = 1 ,ncols = 2)
# The axes is nothing but an array containing axes and we can show it by

for i in axes:
    i.plot(x,y)
```

```python
figs,axes  = plt.subplots(nrows = 1 ,ncols = 2)
# We can also index call off of the axis array

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()

# In here the subplots takes a tuple the fist argument of the tuples creates an figure object and the axes argument
# uses the figure object to craeate a plot 
```

# Figure size and DPI

```python
fig = plt.figure(figsize = (8,2)) # ,dpi = 100)
# dpi stands for dots per inch 
ax  = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

### The above method can also be used in subplots method

```python
figs,axes = plt.subplots(figsize = (8,2))
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2 , ncols = 1 , figsize =(8,2) )
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# How to customize your plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y , color = '#FF8C00')#Name the color or use RGB hex codes
# To change line width use
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# To change line width use
ax.plot(x,y , color = '#FF8C00',linewidth = .5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# to change line opacity 
ax.plot(x,y ,color = 'purple',lw = 3 , alpha = 0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#To change linestyle use the ls or linestyle argument
ax.plot(x,y ,color = 'purple',lw = 3,ls = "-")
```

```python
 fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# to mark the points on the plot we use
ax.plot(x,y,color='Purple',lw =3,ls = '-',marker ='o',markersize = 10,
       markerfacecolor='yellow',markeredgewidth='3',markeredgecolor = 'green')
```

```python
 fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# to mark the points on the plot we use
ax.plot(x,y,color ='purple',lw=1,ls = '--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
